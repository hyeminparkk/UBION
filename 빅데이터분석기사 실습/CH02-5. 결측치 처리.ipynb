## 결측값 특정값으로 대체 - df.fillna()

```python
import pandas as pd

df = pd.read_csv('../datasets/빅분기 실기 데이터/Ex_Missing.csv')
df
```

```python
df_0 = df.fillna(0)
df_0
```

```python
df_missing = df.fillna('missing')
df_missing
```

## 해당 변수의 앞의 값 / 다음 값으로 대체 
- method='ffill' : 앞의 값으로 대체
- method='bfill' : 다음 값으로 대체

```python
df_ffill = df.fillna(method='ffill')
df_ffill
```

```python
df_bfill = df.fillna(method='bfill')
df_bfill
```

## 평균 대체

```python
df_mean = df.fillna(df.mean())
df_mean
```

```python
df_median = df.fillna(df.median())
df_median
```

```python
# 모든 결측치를 salary의 평균으로 대체

df_other_mean = df.fillna(df['salary'].mean())
df_other_mean
```

## 다른 변수 값으로 대체

```python
df1 = df.copy()
```

```python
import numpy as np

# sales의 결측값을 salary 값으로 대체
# sales가 결측이 아니면 원래 값을, 결측값이면 salary 값으로 대체하여 새로운 변수 sales_new 생성
df1['sales_new'] = np.where(pd.notnull(df1['sales'])==True, df['sales'], df['salary'])
df1
```

## 집단 평균값으로 대체

```python
df.groupby('industry').mean()
```

```python
df_group_mean = df.groupby('industry').apply(lambda x: x.fillna(x.mean()))
df_group_mean
```

```python
fill_values = {1:1000, 2:2000}
```

```python
df_group_value = df.groupby('industry').apply(lambda x: x.fillna(fill_values[x.name]))
df_group_value
```

## salary는 보간법 / sales는 평균 / roe는 'missing'으로 대체

```python
missing_fill_val = {'salary':df.salary.interpolate(),
                   'sales':df.sales.mean(),
                   'roe':'missing'}
print(missing_fill_val)
```

```python
df_multi = df.fillna(missing_fill_val)
df_multi
```
