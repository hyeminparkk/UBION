# 데이터 불러오기
- 범주형 변수 : gender, region

```python
import pandas as pd
df = pd.read_csv('../datasets/빅분기 실기 데이터/vote.csv', encoding='utf-8')
df.head(10)
```

```python
X1 = df[['gender', 'region']]
X2 = df[['edu', 'income', 'age', 'score_gov', 'score_progress', 'score_intention', 'vote', 'parties']]
```

# 범주형 변수 One-hot-encoding 변환
- get_dummies() 이용 

```python
X1['gender'] = X1['gender'].replace([1,2], ['male', 'female'])
X1['region'] = X1['region'].replace([1,2,3,4,5], ['Sudo', 'Chungchueng', 'Honam', 'Youngnam', 'Others'])
X1.head(5)
```

```python
X1_dum = pd.get_dummies(X1)
X1_dum.head()
```

# 자료 통합 및 저장

```python
Fvote = pd.concat([X1_dum, X2], axis=1)
Fvote
```
