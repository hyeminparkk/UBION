# 딥러닝 구동하기위한 케라스 함수 불러오기

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

# 데이터 불러오기

```python
data_set = pd.read_csv("./datasets/ThoraricSurgery.csv" , header=None)
```

```python
data_set
```

```python
data_set[17].value_counts()
```

```python
X = data_set.drop([17], axis=1)
y = data_set[[17]]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=11)
```

# 딥러닝 구조 결정

```python
model = Sequential()
model.add(Dense(30, input_dim=17, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

# 딥러닝 실행

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=100, batch_size=10) # batch_size => 1 epoch 돌 때 10개씩 들어가는것
```

# 예측

```python
pred = model.predict(X_test)
```
